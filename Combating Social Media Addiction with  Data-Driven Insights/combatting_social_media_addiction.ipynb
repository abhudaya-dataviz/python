## Advanced Options

Beyond local editing two things are quite important: editing the notebooks in the markdown format and running Jupyter remotely.
The latter matters when we want to run the code on a faster server.
The former matters since Jupyter's native ipynb format stores a lot of auxiliary data that is
irrelevant to the content,
mostly related to how and where the code is run.
This is confusing for Git, making
reviewing contributions very difficult.
Fortunately there is an alternative---native editing in the markdown format.

### Markdown Files in Jupyter

If you wish to contribute to the content of this book, you need to modify the
source file (md file, not ipynb file) on GitHub.
Using the notedown plugin we
can modify notebooks in the md format directly in Jupyter.


First, install the notedown plugin, run the Jupyter Notebook, and load the plugin:

```
pip install d2l-notedown  # You may need to uninstall the original notedown.
jupyter notebook --NotebookApp.contents_manager_class='notedown.NotedownContentsManager'
```

You may also turn on the notedown plugin by default whenever you run the Jupyter Notebook.
First, generate a Jupyter Notebook configuration file (if it has already been generated, you can skip this step).

```
jupyter notebook --generate-config
```

Then, add the following line to the end of the Jupyter Notebook configuration file (for Linux or macOS, usually in the path `~/.jupyter/jupyter_notebook_config.py`):

```
c.NotebookApp.contents_manager_class = 'notedown.NotedownContentsManager'
```

After that, you only need to run the `jupyter notebook` command to turn on the notedown plugin by default.

### Running Jupyter Notebooks on a Remote Server

Sometimes, you may want to run Jupyter notebooks on a remote server and access it through a browser on your local computer. If Linux or macOS is installed on your local machine (Windows can also support this function through third-party software such as PuTTY), you can use port forwarding:

```
ssh myserver -L 8888:localhost:8888
```

The above string `myserver` is the address of the remote server.
Then we can use http://localhost:8888 to access the remote server `myserver` that runs Jupyter notebooks. We will detail on how to run Jupyter notebooks on AWS instances
later in this appendix.

### Timing

We can use the `ExecuteTime` plugin to time the execution of each code cell in Jupyter notebooks.
Use the following commands to install the plugin:

```
pip install jupyter_contrib_nbextensions
jupyter contrib nbextension install --user
jupyter nbextension enable execute_time/ExecuteTime
```

````python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from google.colab import files
df = pd.read_csv('/content/Students Social Media Addiction (1).csv')
````

````python
#1. Data Understanding & Cleaning
df.shape
print(df.info())
print(df.describe())
print(df.isnull().sum())
````

````python
#2. Exploratory Data Analysis (EDA)
# Age, Gender vs Daily Usage
sns.boxplot(data=df, x='Gender', y='Avg_Daily_Usage_Hours')
plt.title("Daily Usage by Gender")
plt.show()
print("Insight: Female students show higher median screen usage with more variability, indicating higher risk.")

sns.scatterplot(data=df,
                x='Age',
                y='Avg_Daily_Usage_Hours',
                hue='Gender',
                style='Gender')
plt.title("Age, Gender vs Daily Usage")
plt.show()
print("Insight: Younger students in the age group of 18 to 21 have high daily usage with females showing higher usage in this age group. While males have low usage and are more in age group of 22 to 24")


#Sleep vs Academic vs Social Interaction
sns.scatterplot(data=df,
                x='Sleep_Hours_Per_Night',
                y='Avg_Daily_Usage_Hours',
                hue='Affects_Academic_Performance',
                style='Affects_Academic_Performance')
plt.title("Sleep vs Academic Performance vs Social Interaction")
plt.show()

print("Insight: Students with poor academic performance cluster around lower sleep and higher social interaction")

#Analyze how addiction varies across demographics
#Country vs addiction score
top_countries = (df.groupby('Country')['Addicted_Score']
                   .mean()
                   .sort_values(ascending=False)
                   .head(10)
                   .reset_index())

sns.barplot(data=top_countries, x='Addicted_Score', y='Country')
plt.title("Top 10 Countries by Addiction Score")
plt.show()
print(top_countries['Country'])
print("Insight: A small number of countries show disproportionately high addiction levels.")


#Age vs addiction score
age_addiction = (df.groupby('Age')['Addicted_Score']
                   .mean()
                   .sort_values(ascending=False)
                   .reset_index())

sns.barplot(data=age_addiction, x='Age', y='Addicted_Score')
plt.title("Addiction Score by Age")
plt.xlabel("Age")
plt.ylabel("Average Addiction Score")
plt.show()

print("Insight: 18 to 21 age groups exhibit higher addiction levels, indicating that younger students may be more vulnerable to excessive usage.")


# Academic Level  vs addiction score
level_addiction = (df.groupby('Academic_Level')['Addicted_Score']
                     .mean()
                     .sort_values(ascending=False)
                     .reset_index())

sns.barplot(data=level_addiction, x='Addicted_Score', y='Academic_Level')
plt.title("Addiction Score by Academic Level")
plt.xlabel("Average Addiction Score")
plt.ylabel("Academic Level")
plt.show()

print("Insight: Students in graduate and undergraduate academic levels show higher addiction risks  possibly due to academic pressure or environment.While highschool have lower addiction score")




````

````python
#3. Aggregation & Insights
agg_gender = df.groupby("Gender")["Addicted_Score"].mean()
agg_age = df.groupby("Age")["Addicted_Score"].mean()
agg_level = df.groupby("Academic_Level")["Addicted_Score"].mean()

print("\nAverage Addiction Score by Gender:\n", agg_gender)
print("\nAverage Addiction Score by Age:\n", agg_age)
print("\nAverage Addiction Score by Academic Level:\n", agg_level)
print("Female has more average addiction score than males. Lower age group has more average addiction score. High school have considerably high addiction score than Graduate and undergraduate")
````

````python
#4. Functions, Loops, and Conditionals

#Classify risk level (Low/Medium/High) based on usage hours

def risk_level(hours):
    if hours < 2:
        return "Low"
    elif hours < 5:
        return "Medium"
    else:
        return "High"

df['Usage_Risk_Level'] = df['Avg_Daily_Usage_Hours'].apply(risk_level)
df[['Avg_Daily_Usage_Hours','Usage_Risk_Level']].head()


#Suggest digital detox strategies using if-else blocks

def detox_suggestion(risk):
    if risk == "High":
        return "Immediate Digital Detox: Limit usage + Seek counseling"
    elif risk == "Medium":
        return "Moderation required: Scheduled breaks & reduced screen time"
    else:
        return "Healthy usage, Maintain Balance"

df['Detox_Strategy'] = df['Usage_Risk_Level'].apply(detox_suggestion)
df[['Usage_Risk_Level','Detox_Strategy']].head()

print(df.head())

````
